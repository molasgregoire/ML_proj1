## 1. Preprocessing

### Create data set we can work on 

```python
# copy dataset
my_tx = tX.copy()

# associate index with feature_name
my_features = np.array([[index, name] for index, name in enumerate (feature_names)])
my_features.shape
```

### Extract PRI_jet_num

```python
# extract PRI_jet_num from my_tx (because it is a classifier feature)
PRI_jet_num = my_tx[:,22]
my_tx = np.c_[my_tx[:,:22], my_tx[:,23:]]

###TEST
PRI_jet_num_test = X_test[:,22]
X_test = np.c_[X_test[:,:22], X_test[:,23:]]

# remove PRI_jet_num from the list of feature
my_features = np.delete(my_features, (22), axis = 0)
```

###  Dealing with missing data

```python
# A look at nan values
mask = (my_tx == -999)
nb_rows = my_tx.shape[0]

emptyRate = 100*np.sum(mask / nb_rows, axis =0)

# Add empty rate to feature description
my_features = np.concatenate((my_features, emptyRate[:,None]), axis = 1)

# Look up for percentage of -999 per column:
my_features
```

We remove feature with more than 70% of missing data, we keep the other and will see how to cope with

```python
# get rid of columns with more than 70% of empty 
mask = np.sum(mask/nb_rows, axis =0) < 0.7

lt70 = my_tx[:, mask]
lt70_features = my_features[mask, :]

###TEST -> apply same mask
X_test_70 = X_test[:, mask]

lt70.shape, lt70_features.shape
```

We create a new feature corresponding to the nbr of values replaced by mean (aka the number of error)

```python
# add a new feature: "# of missing values"
nb_missing = (lt70 == -999).sum(axis = 1)

###TEST
nb_missing_test = (X_test_70 == -999).sum(axis = 1)

print("Among {N} datapoints, there are: {n} datapoints with more than one missing value"
      .format(N = lt70.shape[0], n = np.sum(nb_missing > 0)))
```

```python
#repalce missing values by the median of the feature
replaceByMedian(lt70, -999)

### TEST
replaceByMedian(X_test_70, -999)
```

```python
#standardize the data set (to have a similar scale on each feature)
lt70_stdz = standardize(lt70)

### TEST
X_test_70_stdz = standardize(X_test_70)
```

```python
#Compute correlation
Corr_y = correlation_pearson(lt70_stdz_aug, y)[:-1,-1]
Corr_y_abs = abs(Corr_y)

#Gather tuple of indices of "highly" correlated augmented features and prediction
idList = []
thr = 0.19
for i in Corr_y_abs:
    if i > thr and i < 1:
        idList.append(np.where(Corr_y_abs == i)[0][0])
        print('column: {id_} \n y_correlation: {value} \n Corresponding feature: {feature} \n degree {degree}\n'
              .format(id_= idList[-1],
                      value = i,
                      feature = lt70_features[idList[-1] % nb_feat, :-1],
                      degree = int(idList[-1]/nb_feat)+2))
```
